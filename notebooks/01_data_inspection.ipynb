```python
not_contacted = bank_df[bank_df["pdays"] == 999]
not_contacted["y"].value_counts(normalize=True)
```

```python
contacted = bank_df[bank_df["pdays"] != 999]
contacted["y"].value_counts(normalize=True)
```

```python
contacted["previous"].value_counts( )
```

```python
conversion_contacts = contacted.groupby("previous")["y"].value_counts(normalize=True).reset_index(name="conversion_rate")
conversion_contacts
```
