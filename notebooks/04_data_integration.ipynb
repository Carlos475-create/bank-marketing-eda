```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bank_df = pd.read_csv("../data/processed/bank_clean.csv")
customers_df = pd.read_csv("../data/processed/customer_clean.csv")
```

```python
print(f"Bank dataframe has {bank_df["id_"].nunique()} unique id's and customers dataframe has {customers_df["id_"].nunique()}")
```

```python
set_a = set(bank_df["id_"])
set_b = set(customers_df["id_"])
intersection = set_a.intersection(set_b)
len(intersection)
```

After validating the unique identifiers in both datasets, it was confirmed that there are 42,752 common IDs between the bank campaign dataset and the customer demographic dataset.
Since all campaign records have a corresponding demographic record, an inner join will be performed based on the id_ column.
This ensures that no campaign data is lost during the integration process.

```python
df_integrated = pd.merge(left=bank_df, right=customers_df, on="id_", how="inner")
df_integrated.head()
```

```python
df_integrated.to_csv("../data/processed/integrated.csv")
```
